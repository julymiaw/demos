```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 从CSV文件读取数据
data = pd.read_csv('output.csv', sep='\t', header=None,  index_col=0)
```

```python
# 提取数据列
n = data.loc['n']
F1a = data.loc['F1a']
F1b = data.loc['F1b']
F2a = data.loc['F2a']
F2b = data.loc['F2b']
```

```python
# 创建一个新的图形
plt.figure(figsize=(10, 6))
```

```python
# 绘制曲线
plt.plot(n, F1a, label='F1a')
plt.plot(n, F1b, label='F1b')
plt.plot(n, F2a, label='F2a')
plt.plot(n, F2b, label='F2b')

# 添加标题和标签
plt.rcParams['font.sans-serif'] = 'SimHei'  # 设置使用中文黑体
plt.title('数据可视化')
plt.xlabel('n')
plt.ylabel('值')

# 添加图例
plt.legend()

# 显示图形
plt.show()
```
